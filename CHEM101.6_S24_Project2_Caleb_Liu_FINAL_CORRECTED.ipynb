**Use the full DSSP set of secondary structure assignments to plot the population of another secondary structure type and visualize it with a representative frame or frames in VMD**

**Consult the original DSSP paper (or online documentaiton) to confirm the criteria for this assignment** 

Kabsch W, Sander C (1983). Dictionary of protein secondary structure: pattern recognition of hydrogen-bonded and geometrical features. Biopolymers 22 (12): 2577-637. doi:10.1002/bip.360221211

dssp=md.compute_dssp(trj,simplified=False)

https://mdtraj.org/1.9.4/api/generated/mdtraj.compute_dssp.html

The DSSP assignment codes are:

H : Alpha helix

B : Residue in isolated beta-bridge

E : Extended strand, participates in beta ladder

G : 3-helix (3/10 helix)

I : 5 helix (pi helix)

T : hydrogen bonded turn

S : bend

  : Loops and irregular elements

The simplified DSSP codes are:

H : Helix. Either of the H, G, or I codes.

E : Strand. Either of the E, or B codes.

C : Coil. Either of the T, S or   codes.

```python
####HOMEWORK ######
#Use the full DSSP Definition and plot the population of another secondary structure type 
#https://mdtraj.org/1.9.4/api/generated/mdtraj.compute_dssp.html

dssp_full=md.compute_dssp(trj,simplified=False)
def decompose_simple(dssp_full,simplified_type):
    if simplified_type == 'H':
        types = ['H','G','I']
    elif simplified_type == 'E':
        types = ['E','B']
    elif simplified_type == 'C':
        types = ['T','S',' ']
    else:
        print("Invalid simplified type")
        return
    
    counts = np.empty(len(types), dtype=np.ndarray)
    probE = np.empty(len(types), dtype=np.ndarray)
    for i in range(len(types)):
        counts[i] = np.copy(dssp_full)
        counts[i][counts[i]!=types[i]] = 0
        counts[i][counts[i]==types[i]] = 1
        
        counts[i]=counts[i].astype(int)
        
        probE[i]=np.sum(counts[i], axis=0).astype(float)/len(counts[i])
    
    return probE, types
        
Hprobs, types = decompose_simple(dssp_full,'H')
for i in range(len(Hprobs)):
    plt.plot(residue_num,Hprobs[i], label=types[i])

plt.plot(residue_num,Hprop,'r-.',label='overall helix overlay', alpha=0.5)
plt.xlabel('Residue', size=12)
plt.ylabel('Helix Secondary Strcture Fraction', size=12)
plt.legend()
plt.show()

Eprobs, types = decompose_simple(dssp_full,'E')
for i in range(len(Eprobs)):
    plt.plot(residue_num,Eprobs[i], label=types[i])

plt.plot(residue_num,Eprop,'r-.',label='overall strand overlay', alpha=0.5)
plt.xlabel('Residue', size=12)
plt.ylabel('Strand Secondary Strcture Fraction', size=12)
plt.legend()
plt.show()

Cprobs, types = decompose_simple(dssp_full,'C')
for i in range(len(Cprobs)):
    plt.plot(residue_num,Cprobs[i], label=types[i])
plt.xlabel('Residue', size=12)
plt.ylabel('Coil Secondary Strcture Fraction', size=12)
plt.legend()
plt.show()

#sanity check
total=np.sum(np.vstack((Hprobs[0],Hprobs[1],Hprobs[2],Eprobs[0],Eprobs[1],Cprobs[0],Cprobs[1],Cprobs[2])),axis=0)
plt.plot(residue_num,total)
```

```python
#### Challenge Problem Write a Function to Compute the Population of Any Secondary Structure Type
#### Write a loop to plot all populations
#### Write code to output the residue numbers of every residue with > 90% each SS type

def list_90_percent(dssp_full,types):
    #array will store values according to the documentation (ex. idx0='A')
    probE = np.empty(len(types), dtype=np.ndarray)
    storage = np.empty(len(types), dtype=np.ndarray)
    counts = np.empty(len(types), dtype=np.ndarray)
    total_frames = trj.n_frames
    
    for i in range(len(types)):
        counts[i] = np.copy(dssp_full)
        counts[i][counts[i]!=types[i]] = 0
        counts[i][counts[i]==types[i]] = 1
        
        counts[i]=counts[i].astype(int)
        
        probE[i]=np.sum(counts[i], axis=0).astype(float)/total_frames
        
        
        
        for j in range(len(probE[i])):
            if probE[i][j] > 0.9:
                storage[i] = np.append(storage[i], j+242)
        
        storage[i] = np.delete(storage[i], 0)
        
    return storage

types = ['H','G','I','E','B','T','S',' ']
storage = list_90_percent(dssp_full,types)
for i in range(len(storage)):
    print("Residues with more than 90% population of " + types[i] + ": " + str(storage[i]))
```

```python
def calc_phipsi_FES(res):
    free_energy,xedges,yedges=np.histogram2d(dihedral_dict['psi'][res],dihedral_dict['phi'][res],36,[[-3.14,3.14],[-3.14,3.14]],density=True,weights=None)
    free_energy=np.log(np.flipud(free_energy)+.000001)
    kbT=300*0.008314/4.184 # kT in kcal/mol
    dG=-(kbT)*free_energy
    return dG

from collections import deque #using a FIFO queue

#let's find a residue without guess-and-check with a growing region finder
list_multibasin_residues = [] #which residues have multiple basins
all_basin_indices = {} #store every basin by residue
multibasin_basin_indices = {} #store only multibasin residues

####################################################
cutoff_free_energy = 1 #this can be altered
####################################################

#basin finder iterates over every residue
for res in range(len(residue_id)-2):#-2 needed due to a quirk in residue_id
    print("currently checking: RESID" + str(res+243))
    energy_profile = calc_phipsi_FES(res+243) #get the histogram
    list_checked_coords = [] #list of checked coordinates so we know what to include and exclude from our basins
    check_queue = deque([]) #FIFO queue
    num_basins = 0
    residue_basins=[] #basins on each residue
    for x in range(np.shape(energy_profile)[0]): #iterate through every index
        for y in range(np.shape(energy_profile)[1]): #iterate through every index
            if (energy_profile[x,y] < cutoff_free_energy) & ([x,y] not in list_checked_coords): #new point in basin cutoff
                print("basin found" + str(x) + ", " + str(y))
                num_basins+=1
                check_queue.append([x,y]) #add to fifo queue
                
                basin_indices=[]
                
                while len(check_queue) != 0:
                    curr_checked = check_queue.popleft()
                    curr_x = curr_checked[0]
                    curr_y = curr_checked[1]
                    
                    if (curr_x < 0) | (curr_x >= (np.shape(energy_profile)[0])): #check for out-of-bounds
                        continue
                    if (curr_y < 0) | (curr_y >= (np.shape(energy_profile)[1])):
                        continue
                    
                    #if the checked point is not checked and is also in the basin, add it to the list and grow the region
                    if (energy_profile[curr_x,curr_y] < cutoff_free_energy) & ([curr_x, curr_y] not in list_checked_coords):
                        check_queue.append([curr_x-1,curr_y-1])
                        check_queue.append([curr_x-1,curr_y])
                        check_queue.append([curr_x-1,curr_y+1])
                        check_queue.append([curr_x,curr_y-1])
                        check_queue.append([curr_x,curr_y+1])
                        check_queue.append([curr_x+1,curr_y-1])
                        check_queue.append([curr_x+1,curr_y])
                        check_queue.append([curr_x+1,curr_y+1])
                    
                        list_checked_coords.append([curr_x, curr_y])
                        basin_indices.append([curr_x,curr_y])
                
                residue_basins.append(basin_indices)
                
            else:
                list_checked_coords.append([x,y])
    
    
    
    if num_basins > 1:
        list_multibasin_residues.append(res+243)
        res_basin_indices={res+243:residue_basins}
        multibasin_basin_indices.update(res_basin_indices)
        all_basin_indices.update(res_basin_indices)
    
    else:
        res_basin_indices={res+243:residue_basins} #keeps dimensions right
        all_basin_indices.update(res_basin_indices)

#Due to the nature of this algorithm, basins that cross the -pi/pi boundary are counted. These can be simply plotted and checked manually as there aren't a lot fo them
list_multibasin_residues = np.asarray(list_multibasin_residues)

print("Residues with multiple deep basins: ", list_multibasin_residues)

flag = False
while (flag == False):
    desired_basin=int(input("Specify desired basin (screenshots are for residue 400)"))
    if desired_basin in multibasin_basin_indices.keys():
        print("Basin indices: ", multibasin_basin_indices[desired_basin])
        plot_phipsi_FES(desired_basin)
        flag = True
    else: 
        print("Basin not found")

basin1=np.asarray(Image.open('/ipynb_files/CHEM96_06/Project_2/photos/Basin1phipsi.png'))
basin2=np.asarray(Image.open('/ipynb_files/CHEM96_06/Project_2/photos/Basin2phipsi.png'))

fig, axs = plt.subplots(1,2)
fig.set_size_inches(18.5,10)
axs[0].axis('off')
axs[1].axis('off')
axs[0].title.set_text('basin1')
axs[1].title.set_text('basin2')
axs[0].imshow(basin1)
axs[1].imshow(basin2)

print("As this residue is on the outside of the protein and is in contact with the solvent, it is likely that the configurational")
print("space is stochastically driven.")
```
